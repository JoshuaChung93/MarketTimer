```python
code = 'KQ11'
df = fdr.DataReader('KQ11','2022-08').reset_index()
# 3,5,10 이동평균 딕셔너리에 할당
df['close_sma3d'] = df['Close'].rolling(3).mean()
df['close_sma5d'] = df['Close'].rolling(5).mean()
df['close_sma10d'] = df['Close'].rolling(10).mean()
```

```python
df.iloc[-10:]
```

```python
# dataframe 재구성
df2 = df.loc[: ,['Date','Close', 'close_sma3d','close_sma5d','close_sma10d']].iloc[-20:]
df2
```

```python
print('\n--- Close for 2022-08-03 ---\n')

print(df.loc[ df['Date'] == '2022-08-03', 'Close'])

print('\n--- Previous Close < Close ---\n')

df['Previous Close'] = df['Close'].shift(1)
# df['Previous Date']  = df['Date'].shift(1)

lower_close = df[ df['Previous Close'] < df['Close'] ]
print(lower_close[['Previous Close', 'Close', 'Date']])

print('\n--- alert_three,five,ten_days ---\n')

# 3, 5, 10 마켓타이밍 유무 딕셔너리에 할당
alert_three_days = (df2['Close'] > df2['close_sma3d'] )
alert_five_days = (df2['Close'] > df2['close_sma5d'] )
alert_ten_days = (df2['Close'] > df2['close_sma10d'] )
df2[3] = alert_three_days
df2[5] = alert_five_days
df2[10] = alert_ten_days
print(df2)

#def display(row):
#    print(f" - {row['Date']} Signal 발생! now_price {row['Close']} 3일이동평균 {row['close_sma3d']}")
#alert_three_days.apply(display, axis=1)

for index, row in (df2[3] | df2[5] | df2[10]).iterrows():
    print(f" - {row['Date']} Signal 발생! now_price {row['Close']} 3일이동평균 {row['close_sma3d']}")
    print(f" - {row['Date']} Signal 발생! now_price {row['Close']} 5일이동평균 {row['close_sma5d']}")
    print(f" - {row['Date']} Signal 발생! now_price {row['Close']} 10일이동평균 {row['close_sma10d']}")
```
